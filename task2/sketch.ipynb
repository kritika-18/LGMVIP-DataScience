# Task 2: Image to Pencil Sketch with Python

## Importing necesary Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

## Reading the image

```python
loc='C:/Users/tdeor/Downloads/'
name='flower.jpeg'
image=cv2.imread(loc+name)
```

## Displaying image using OpenCV

```python
cv2.imshow('Original image' ,image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Displaying image using Matplotliib

```python
plt.imshow(image)
plt.axis(False)
plt.show()
```

## Converting BGR to RGB

```python
rgb_img=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.axis(False)
plt.show()
```

## Converting image to Gray Image

```python
gray_img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

## Inverting the image

```python
invert_img=cv2.bitwise_not(gray_img)
```

## Blurring the image

```python
blurr_img=cv2.GaussianBlur(invert_img,(101,101),0)
```

## Inverting the Blurred Image

```python
inv_blurr_img=cv2.bitwise_not(blurr_img)
```

## Converting to sketch

```python
sketch=cv2.divide(gray_img,inv_blurr_img, scale=255.0)
```

## Saving the sketch

```python
cv2.imwrite("Sketch.jpeg", sketch)
```

## Displaying the sketch

```python
cv2.imshow("Sketch",sketch)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Dispalying both the images side by side

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("ORIGINAL IMAGE",size=15)
plt.imshow(rgb_img)
plt.axis(False)

plt.subplot(1,2,2)
plt.title("SKETCH",size=15)
rgb_sketch=cv2.cvtColor(sketch,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis(False)

plt.show()
```
